```python
import requests
import pandas as pd
import xmltodict
import json

appId="69ea111ae1ce914c5549e74c8c873b711d5e4efe"
url="https://api.e-stat.go.jp/rest/3.0/app/getStatsData?cdCat01=A1101%2CA110101%2CA110102&appId="+str(appId)+"&lang=J&statsDataId=0000010101&metaGetFlg=Y&cntGetFlg=N&explanationGetFlg=Y&annotationGetFlg=Y&sectionHeaderFlg=1&replaceSpChars=0"
print(url)

x=requests.get(url)
dict=xmltodict.parse(x.content) #type
js=json.dumps(dict,indent=4,ensure_ascii=False) #Unicode(ensure_ascii=False)
print(type(js))
```

```python
#json_load = json.loads(js) #
json_load = eval(js)
print(type(json_load))
```

```python
json_val  = json_load['GET_STATS_DATA']['STATISTICAL_DATA']['CLASS_INF']['CLASS_OBJ']
type(type(json_val))
#json_val
```

```python
json_val1=json_val[2]['CLASS']
#type(json_val1)
json_val1
```

```python
result=[]
for d in json_val1:
    code=d['@code']
    name=d['@name']
    print(code+":"+name)
    result.append([code,name])
result
```

```python
print(f"json_val={type(json_val1)}")
print(f"result  ={type(result)}")
```
