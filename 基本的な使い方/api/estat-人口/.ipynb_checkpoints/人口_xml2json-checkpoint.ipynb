```python
import requests
import pandas as pd
import xmltodict
import json

appId="69ea111ae1ce914c5549e74c8c873b711d5e4efe"
url="https://api.e-stat.go.jp/rest/3.0/app/getStatsData?cdCat01=A1101%2CA110101%2CA110102&appId="+str(appId)+"&lang=J&statsDataId=0000010101&metaGetFlg=Y&cntGetFlg=N&explanationGetFlg=Y&annotationGetFlg=Y&sectionHeaderFlg=1&replaceSpChars=0"
print(url)

x=requests.get(url)
dict=xmltodict.parse(x.content) #type変更

js=json.dumps(dict,indent=4,ensure_ascii=False) #Unicodeエスケープされた文字列を通常の文字列に変換(ensure_ascii=False)
json_load = eval(js)

json_val  = json_load['GET_STATS_DATA']
json_val
```

```python
json_val2=json_val['STATISTICAL_DATA']['DATA_INF']['VALUE']
json_val2
```

```python
c='A1101'
f='2022100000'
d={}
for v in json_val2:
    if v['@cat01']==c and v['@time']==f:
        d[v['@area']]=v['#text']
d
```
