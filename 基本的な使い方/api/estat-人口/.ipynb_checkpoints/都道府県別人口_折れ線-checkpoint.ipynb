```python
import requests
import xmltodict
import json
import PySimpleGUI as sg
import matplotlib.pyplot as plt

def read_population():
    appId="69ea111ae1ce914c5549e74c8c873b711d5e4efe"
    url="https://api.e-stat.go.jp/rest/3.0/app/getStatsData?cdCat01=A1101%2CA110101%2CA110102&" \
        +"appId="+str(appId)+"&lang=J&statsDataId=0000010101&metaGetFlg=Y&cntGetFlg=N&" \
        +"explanationGetFlg=Y&annotationGetFlg=Y&sectionHeaderFlg=1&replaceSpChars=0"

    x=requests.get(url)
    dict=xmltodict.parse(x.content) #type変更

    js=json.dumps(dict,indent=4,ensure_ascii=False) #Unicodeエスケープされた文字列を通常の文字列に変換(ensure_ascii=False)
    json_load = eval(js)

    json_val1  = json_load['GET_STATS_DATA']['STATISTICAL_DATA']['CLASS_INF']['CLASS_OBJ'][3]['CLASS']
    dict_y=[]
    for v in json_val1:
        dict_y.append((v['@code']))

    json_val2  = json_load['GET_STATS_DATA']['STATISTICAL_DATA']['DATA_INF']['VALUE']
    check1='A1101'
    check2='A110101'
    check3='A110102'
    dict_data_00={}
    dict_data_01={}
    dict_data_02={}
    for v in json_val2:
        if v['@area']=='00000':
            if v['@cat01']==check1:
                dict_data_00[v['@time']]=v['#text']
            if v['@cat01']==check2:
                dict_data_01[v['@time']]=v['#text']
            if v['@cat01']==check3:
                dict_data_02[v['@time']]=v['#text']
    result=[]
    for row in dict_y:
        time=row
        data1=dict_data_00[row]
        data2=dict_data_01[row]
        data3=dict_data_02[row]
        result.append([time,data1,data2,data3])
    return result
    
if __name__ == "__main__":
    data=read_population()
    #print(data)
    x=[(r[0]) for r in data]
    y=[int(r[1]) for r in data]

    plt.plot(x, y)
    plt.show()

#    print(data)
    
```
