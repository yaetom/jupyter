```python
import pandas as pd
import json
import xml.etree.ElementTree as ET
import PySimpleGUI as sg

def read_population():
    #xmlデータを読み込みます
    tree = ET.parse('getStatsData.xml')
    #一番上の階層の要素を取り出します
    root = tree.getroot()
    
    result=[]
    for child in root[2][2][2]:
        #print(child.attrib["code"]+"="+child.attrib["name"])
        code=child.attrib["code"]
        name=child.attrib["name"]
        result.append([code,name])
    
    return result
    
def show_table(data):
    layout=[
        [sg.Table(values=data,
                  headings=["No","都道府県"],
                  auto_size_columns=True,
                  expand_x=True,
                  expand_y=True,
                  justification="right",
                  font=("Arial",14))],
        [sg.Button("閉じる")]]

    window=sg.Window("都道府県番号",layout,size=(600,400))
    while True:
        event,_=window.read()
        if event in ["閉じる",sg.WIN_CLOSED]:
            break
        window.close()

if __name__ == "__main__":
    data=read_population()
    show_table(data)
```
