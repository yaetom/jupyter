```python
import urllib

url="https://app.rakuten.co.jp/services/api/IchibaItem/Search/20220601"
format="json"
applicationid="1057585233871214360"
keyword="うさまる"

url2=url+"?"+"format="+format+"&keyword="+urllib.parse.quote(keyword)+"&applicationId="+applicationid
print(url2)
```

```python
import urllib.request as request
import json

try:
    response=request.urlopen(url2)
    contents=response.read()
    print(contents.decode())
    response.close()
except urllib.error.HTTPError:
    print("not available")

else:
    pass
```

```python
import pandas as pd
import json

temp2=json.loads(contents.decode())
temp3=pd.json_normalize(temp2['Items'])
temp3
```

```python
import urllib.request as request
import json
import pandas as pd
import time

usamaru_list=pd.DataFrame()

try:
    response=request.urlopen(url2)
    contents=response.read()
    response.close()
    temp=json.loads(contents.decode())
    temp10=pd.json_normalize(temp)
    numofpages=temp10.pageCount
    print("取得するページ数",numofpages[0])

    for i in range(numofpages[0]):
        print(i+1,"ページ目")
        time.sleep(10)
        url3=url2+"&pege="+str(i+1)
        response2=request.urlopen(url3)
        contents=response2.read()
        response2.close()
        temp20=json.loads(contents.decode())
        temp21=pd.json_normalize(temp20['Items'])
        usamaru_list=pd.concat([usamaru_list,temp21],axis=0)
except urllib.error.HTTPError:
    print("not available")
else:
    pass
```
