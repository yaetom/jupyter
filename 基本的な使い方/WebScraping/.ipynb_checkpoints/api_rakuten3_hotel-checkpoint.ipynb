```python
import urllib

url="https://app.rakuten.co.jp/services/api/Travel/SimpleHotelSearch/20170426"
format="json"
latitude="128440.51"
longitude="503172.21"
serchRadius="3"
applicationid="1057585233871214360"

url3=url+"?"+"applicationId="+applicationid+"&format="+format+"&latitude="+latitude+"&longitude="+longitude+"&searchRadius="+serchRadius
print(url3)
```

```python
import urllib.request as request

try:
    response=request.urlopen(url3)
    contents=response.read()
    print(contents.decode())
    response.close()
except urllib.error.HTTPError:
    print("not available")
else:
    pass
```

```python
import pandas as pd
import json

temp2=json.loads(contents.decode())

temp4=pd.json_normalize(temp2['pagingInfo'])
temp3=pd.json_normalize(temp2,["hotels","hotel"])
```
