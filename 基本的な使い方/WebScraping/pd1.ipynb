```python
import pandas as pd
import time

url1="https://www.soc.hit-u.ac.jp/research/archives/master.cgi?choice=list&yearID="

list=pd.DataFrame()
for i in range(28,44):
    url2=url1+str(i)
    shuron=pd.read_html(url2)
    shuron[0]["year"]=str(i+1979)
    list=pd.concat([list,shuron[0]],axis=0)
    time.sleep(5)
    print(i+1979,"年完了")
```

```python
new_column_names = ['研究分野', '論文タイトル', '指導教員名','year']
list=list.set_axis(new_column_names, axis=1, inplace=False)
#print(df.index[0])
list.drop(list.index[0],axis=0)
list
```
