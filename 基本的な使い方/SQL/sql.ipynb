```python
import os
import pymysql
import pandas as pd
import matplotlib.pyplot as plt

host='localhost'
port='3306'
user='user001'
password='mysqluser001'
databese='gbizinfo'

conn=pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=databese,
    charset='utf8mb4')
```
