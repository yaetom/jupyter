```python
from bs4 import BeautifulSoup
import requests
import bs4
import pandas as pd

url = "https://it-trend.jp/words#word1"
res = requests.get(url)

parse_html = BeautifulSoup(res.text, "html.parser")
title_lists = parse_html.find_all("a")

#print(title_lists)

#name_list=[]
url_list=[]
alt_list=[]
#prefix="ITトレンド用語："

for i in title_lists:
    if 'href' in i.attrs:
        if 'alt' in i.attrs:
            #aタグのhref属性値を抽出する
#            name_list.append(i.string)
            url_list.append(i.attrs["href"])
            
            alt=i.attrs["alt"]
#            if alt.startswith(prefix):
#                alt = alt[len(prefix):]
            alt_list.append(alt[9:])

url = "https://it-trend.jp/words/identity"
res = requests.get(url)

soup = BeautifulSoup(res.text, "html.parser")

article_body = soup.find("td", class_="first")
text = article_body.get_text()
parts = text.split("/")
print(parts[0][:-1])
print(parts[1][1:])

article_body = soup.find(id="detail_sentence")
print(article_body.get_text())



#print(name_list)
#df_title_url = pd.DataFrame({"Title":name_list,"URL":url_list,"ALT":alt_list})
df_title_url = pd.DataFrame({"URL":url_list,"ALT":alt_list})
df_title_url
```
