```python
## 基本構文

a = [1, 2, "Hello"]

for j in a:
  print(j)
```

```python
## rangeの使い方

for i in range(4):
  print(i)
```

```python
list(range(4))
```

```python
## 辞書とfor文

a = {
    "key1": "Hello",
    "key2": "Bye",
    "num": 1
  }

for i, j in a.items():
  print(i)
  print(j)
```

```python
## enumerateの使い方

a = [1, 2, "Hello"]

for i, j in enumerate(a):
  print(i)
  print(j)
```

```python
## list内包表記の使い方

a = [1, 2, 3]
b = []

for i in a:
  b.append(i * 2)
b
```

```python
[i*2 for i in a]
```

```python
## breakの使い方

a = [1, 2, 3, 4, 5]

for i in a:
  if i > 3:
    break
  print(i)
```
