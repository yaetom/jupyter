```python
age = 20

if age < 12:
  status = "child"
else:
  status = "adult"

status
```

```python
age = 70

if age < 12:
  status = "child"
elif age < 65:
  status = "adult"
else:
  status = "senior"

status
```

```python
age = 60
gender = "female"

if age < 12:
  status = "child"
elif age < 65 and gender == "male":
  status = "male-adult"
elif age < 65 and gender == "female":
  status = "female-adult"
else:
  status = "senior"

status
```
