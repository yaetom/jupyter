```python
## 独自関数

age = 60
gender = "female"

if age < 12:
  status = "child"
elif age < 65 and gender == "male":
  status = "male-adult"
elif age < 65 and gender == "female":
  status = "female-adult"
else:
  status = "senior"

print(status)


age = 70
gender = "male"

if age < 12:
  status = "child"
elif age < 65 and gender == "male":
  status = "male-adult"
elif age < 65 and gender == "female":
  status = "female-adult"
else:
  status = "senior"

print(status)
```

```python
# def 関数名(引数):
#   処理

def ticket(age, gender):
  if age < 12:
    status = "child"
  elif age < 65 and gender == "male":
    status = "male-adult"
  elif age < 65 and gender == "female":
    status = "female-adult"
  else:
    status = "senior"
  return status

```

```python
## 独自関数

age = 60
gender = "female"

status = ticket(age, gender)
print(status)


age = 70
gender = "male"

status = ticket(age, gender)
print(status)
```
