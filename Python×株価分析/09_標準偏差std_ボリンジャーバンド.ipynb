```python
import yfinance as yf
import matplotlib.pyplot as plt

ticker="AMZN" #amazon
df = yf.download(ticker, start="2020-01-01",end="2022-12-31", interval="1d")
df.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

price = df["Close"]
span = 30

df["MA_for_30_days"]=price.rolling(window=span,min_periods=1).mean() #min_periods=1　1日でもデータが入るようにする
df["std"]=price.rolling(window=span,min_periods=1).std() #標準偏差

df["-1σ"] = df["MA_for_30_days"] - 1*df["std"]
df["+1σ"] = df["MA_for_30_days"] + 1*df["std"]

plt.figure(figsize=(15,5))
plt.title(ticker + "trend analysis")
plt.plot(df["Close"], color="blue")
plt.plot(df["-1σ"], color="red")
plt.plot(df["+1σ"], color="red")
plt.fill_between(df.index,df["-1σ"],df["+1σ"],color="red",alpha=0.5)
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend(["Close","-1σ","+1σ"],loc="upper right")
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

ticker="AMZN" #amazon
df = yf.download(ticker, start="2020-01-01",end="2022-12-31", interval="1d")
df.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

price = df["Close"]
span = 30

df["MA_for_30_days"]=price.rolling(window=span,min_periods=1).mean() #min_periods=1　1日でもデータが入るようにする
df["std"]=price.rolling(window=span,min_periods=1).std() #標準偏差

df["-1σ"] = df["MA_for_30_days"] - 1*df["std"]
df["+1σ"] = df["MA_for_30_days"] + 1*df["std"]
df["-2σ"] = df["MA_for_30_days"] - 2*df["std"]
df["+2σ"] = df["MA_for_30_days"] + 2*df["std"]

plt.figure(figsize=(15,5))
plt.title(ticker + "trend analysis")
plt.plot(df["Close"], color="blue")
plt.plot(df["-1σ"], color="red")
plt.plot(df["+1σ"], color="red")
plt.plot(df["-2σ"], color="green")
plt.plot(df["+2σ"], color="green")

plt.fill_between(df.index,df["-1σ"],df["+1σ"],color="red",alpha=0.5)
plt.fill_between(df.index,df["-2σ"],df["+2σ"],color="lightgreen",alpha=0.5)
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend(["Close","-1σ","+1σ","-2σ","+2σ"],loc="upper right")
plt.show()
```
