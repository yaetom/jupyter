```python
import yfinance as yf
```

```python
ticker="AMZN" #amazon

data = yf.download(ticker, start="2022-12-01",end="2022-12-31",period="1y", interval="1d")
data.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
data
```

```python
import matplotlib.pyplot as plt

df = data
df = df.drop("Volume",axis=1) #Volume除外

plt.figure(figsize=(15,5))
plt.plot(df.index,df)
plt.legend(df) #凡例
plt.show()
```

```python
!pip install mplfinance
```

```python
import mplfinance as mpf
mpf.plot(df,type="candle",figratio=(15,5))
```

```python
mpf.plot(df,type="candle",figratio=(15,5),style="yahoo")
```

```python
mpf.plot(df,type="candle",figratio=(15,5),style="yahoo",mav=3) #mav:移動平均線
```
