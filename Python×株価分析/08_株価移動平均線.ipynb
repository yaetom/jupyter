```python
import yfinance as yf
import matplotlib.pyplot as plt

ticker="AMZN" #amazon
df = yf.download(ticker, start="2020-01-01",end="2022-12-31", interval="1d")
df.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

price = df["Close"]
span = 30

df["MA_for_30_days"]=price.rolling(window=span,min_periods=1).mean() #min_periods=1　1日でもデータが入るようにする
df_mb = df[["Close","MA_for_30_days"]]

plt.figure(figsize=(15,5))
plt.plot(df_mb.index,df_mb)
plt.legend(df_mb)
plt.show()
```
