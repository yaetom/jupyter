```python
import yfinance as yf
```

```python
ticker="AMZN" #amazon

data = yf.download(ticker, start="2022-12-01",end="2022-12-31",period="1y", interval="1d")
data.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
data
```

```python
import matplotlib.pyplot as plt

df = data
df = df.drop("Volume",axis=1) #Volume

plt.figure(figsize=(15,5))
plt.plot(df.index,df)
plt.legend(df) #
plt.show()
```

```python
ticker="AMZN" #amazon
df = yf.download(ticker, start="2020-01-01",end="2022-12-31", interval="1d")
df.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

price = df["Close"]
span = 5

df["MA_for_5_days"]=price.rolling(window=span,min_periods=1).mean() #min_periods=11
```

```python
df #
```

```python
df_ma = df[["Close","MA_for_5_days"]]

plt.figure(figsize=(15,5))
plt.plot(df_ma.index,df_ma)
plt.legend(df_ma)
plt.show()
```

```python
span = 30

df["MA_for_30_days"]=price.rolling(window=span,min_periods=1).mean() #min_periods=11
df_mb = df[["Close","MA_for_30_days"]]

plt.figure(figsize=(15,5))
plt.plot(df_mb.index,df_mb)
plt.legend(df_mb)
plt.show()
```
