```python
df = df.reset_index()
df_a = df[["index","survived"]]
df_b = df[["index","pclass"]]
```

```python
pd.merge(df_a,df_b,on="index",how="inner") #結合 innerは双方があるもののみ結合
```

```python
pd.concat([df_a,df_b],axis=1) #横方向に結合
```

```python
pd.concat([df_a,df_b],axis=0) #縦方向に結合
```
