```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy import integrate
%matplotlib inline
```

```python
plt.style.use("ggplot")
```

```python
x = np.arange(-100,100,0.01)
y = np.exp(-x**2) #ガウス関数
```

```python
plt.plot(x,y)
plt.xlim([-5,5]) #ガウス関数
```

```python
integrate.simps(y,dx=0.01)
```

$\int_{-\infty}^{\infty}\exp(-ax^2)dx=\sqrt{\frac{\pi}{a}}$ 

```python
np.sqrt(np.pi)
```

```python
integrate.cumtrapz(y,dx=0.01) #累積和
```
