```python
import numpy as np
import scipy as sp
import matplotlib.pylab as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
plt.style.use("ggplot")
np.random.seed(123)
```

```python
x = np.arange(20)
y = 2.0 * x
y_noise = np.random.randn(20)
y = y + y_noise
```

```python
plt.scatter(x,y)
```

```python
param = sp.polyfit(x,y,1) #多項式近似 1次
```

```python
param
```

```python
plt.plot(x,param[0]*x+param[1])
plt.scatter(x,y)
```

```python
x2 = np.arange(15)
y2 = np.exp(0.4 * x2)
y2 = y2 + 10.0 * np.random.randn(15)
```

```python
plt.scatter(x2,y2)
```

```python
def nonlinear_fit_(x,a,b,c):
    return a * np.exp(b * x) + c
```

```python
popt, pcov = curve_fit(nonlinear_fit,x2,y2)
```

```python
popt
```

```python
plt.scatter(x2,y2)
plt.plot(x2, nonlinear_fit_(x2, popt[0], popt[1], popt[2]))
```
