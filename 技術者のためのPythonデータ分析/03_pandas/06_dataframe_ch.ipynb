```python
df=pd.read_excel("./../99_resources/physical_measurement2.xlsx",index_col=0)
```

```python
df2=df.sort_index()
```

```python
df2.dropna() #NaNを削除
```

```python
df2.fillna(value=1.0) #NaNを指定したvalueとする
```

```python
df4=df3.drop_duplicates() #重複を削除
```
