```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("ggplot")
```

```python
df=pd.read_excel("./../99_resources/multiple_data 2.xlsx")
```

```python
df.head()
```

```python
groups = df.groupby("param")
```

```python
for name,group in groups:
    plt.scatter(group["X"],group["Y"],label=name)
plt.legend()
```

```python
df_1 = groups.get_group(2)
plt.plot(df_1['X'], df_1['Y'])
plt.show()
```
