```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
a = np.arange(10)
```

```python
a
```

```python
a.cumsum() #キムサム　累積和
```

```python
x = np.random.randn(100).cumsum() #乱数の値が次々に足されていくことで、ブラウン運動していく
y = np.random.randn(100).cumsum() #乱数の値が次々に足されていくことで、ブラウン運動していく
z = np.random.randn(100).cumsum() #乱数の値が次々に足されていくことで、ブラウン運動していく
```

```python
%matplotlib inline
#%matplotlib notebook
fig = plt.figure() #figureを作る枠をつくる

#ax = Axes3D(fig) #figが3Dに変換される
ax = fig.add_subplot(111, projection='3d')

#fig.add_axes(ax)
ax.plot(x,y,z)
plt.show()
```

```python
%matplotlib inline

fig = plt.figure()
ax = Axes3D(fig)
ax.plot(x,y,z)

fig.add_axes(ax)
plt.show()
```
