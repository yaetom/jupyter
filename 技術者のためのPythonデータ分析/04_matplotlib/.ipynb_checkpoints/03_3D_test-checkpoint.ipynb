```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = np.random.randn(100).cumsum() #乱数の値が次々に足されていくことで、ブラウン運動していく
y = np.random.randn(100).cumsum() #乱数の値が次々に足されていくことで、ブラウン運動していく
z = np.random.randn(100).cumsum() #乱数の値が次々に足されていくことで、ブラウン運動していく

%matplotlib notebook
subplot3d = plt.subplot(111,projection="3d")

subplot3d.scatter(x,y,z)
subplot3d.set_zlim3d([0,9])
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = np.random.randn(100).cumsum() #乱数の値が次々に足されていくことで、ブラウン運動していく
y = np.random.randn(100).cumsum() #乱数の値が次々に足されていくことで、ブラウン運動していく
z = np.random.randn(100).cumsum() #乱数の値が次々に足されていくことで、ブラウン運動していく

%matplotlib inline
#%matplotlib notebook
fig = plt.figure()
ax = Axes3D(fig)
ax.plot(x,y,z)
fig.add_axes(ax)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = np.random.randn(100).cumsum()
y = np.random.randn(100).cumsum()
z = np.random.randn(100).cumsum()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # 3Dプロジェクションを設定
ax.plot(x, y, z)
plt.show()
```
