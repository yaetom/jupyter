```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("./../99_resources/Concentration.xlsx",index_col=0)
```

```python
df.head()
```

```python
plt.scatter(df.index,df["C[mol/L]"]) #scatter = 散布図
plt.xlabel("time[s]",fontsize=16)
plt.ylabel("C[mol/L]",fontsize=16)
```

```python
plt.plot(df.index,df["C[mol/L]"]) #plot = 折れ線グラフ
```

```python
plt.bar(df.index,df["C[mol/L]"]) #bar = 棒グラフ
```

```python
x = np.arange(0,15,0.01)
y= np.exp(-0.1*x)
plt.bar(df.index,df["C[mol/L]"])
plt.plot(x,y,color="red")
```
