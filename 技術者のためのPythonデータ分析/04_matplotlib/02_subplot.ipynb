```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
x = np.arange(-10,10,0.01)
y = 1.0 / (1.0 + np.exp(-x)) #exp : エクスポネンシャル
```

```python
x2 = np.arange(-6,6,0.01)
y2 = np.sin(x2)
```

```python
plt.subplot(2,1,1) #2行1列の画面、1を使用
plt.plot(x,y)
plt.subplot(2,1,2) #2を使用
plt.plot(x2,y2)
```

```python
plt.subplot(2,2,1) #2行1列の画面、1を使用
plt.plot(x,y)
plt.subplot(2,2,2) #2を使用
plt.plot(x2,y2)
```

```python
plt.subplot(2,2,1) #2行1列の画面、1を使用
plt.plot(x,y)
plt.subplot(2,2,4) #4を使用
plt.plot(x2,y2)
```
